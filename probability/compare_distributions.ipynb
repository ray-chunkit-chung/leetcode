```python
import numpy as np
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt
```

```python
# Generate two example datasets (you can replace these with your own data)
np.random.seed(42)
sample1 = np.random.normal(loc=0, scale=1, size=1000)
sample2 = np.random.normal(loc=0.5, scale=1, size=1000)

# Perform the KS test
ks_statistic, p_value = ks_2samp(sample1, sample2)

# Plot the cumulative distribution functions (CDFs)
plt.figure(figsize=(8, 6))
plt.plot(np.sort(sample1), np.arange(1, len(sample1) + 1) / len(sample1), label="Sample 1")
plt.plot(np.sort(sample2), np.arange(1, len(sample2) + 1) / len(sample2), label="Sample 2")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("CDFs of Sample 1 and Sample 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Print the KS test results
print(f"KS Statistic: {ks_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The distributions are different.")
else:
    print("Fail to reject the null hypothesis: The distributions are similar.")
```
